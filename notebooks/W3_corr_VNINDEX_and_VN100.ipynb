```python
# Caculate corralation between VNINDEX and VN100
companys = get_VN100_company_names()
for company in companys:
    company_df = get_return_rates(df,company)
    if not company_df.empty :
        print(f'Correlation indexs between VNINDEX and {company} are:')
        compute_correlation(VNINDEX_first_df,company_df,'daily_returns,weekly_returns,monthly_returns')
        print('\n')
```
